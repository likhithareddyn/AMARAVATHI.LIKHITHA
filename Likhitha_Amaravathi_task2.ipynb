```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Diwali Sales Data.csv',encoding='ISO-8859-1')
```

```python
# 1. Plot the count of males and females in the dataset
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Gender')
plt.title('Count of Males and Females')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# 2. Give the sum of amounts spent by each gender and plot the corresponding graph
gender_spending = df.groupby('Gender')['Amount'].sum()
plt.figure(figsize=(8, 6))
gender_spending.plot(kind='bar')
plt.title('Total Amount Spent by Each Gender')
plt.xlabel('Gender')
plt.ylabel('Total Amount Spent')
plt.show()
```

```python
# 3. Count each age group and provide individual counts grouped by gender
age_gender_counts = df.groupby(['Age Group', 'Gender']).size().unstack()
print(age_gender_counts)
```

```python
# 4. Plot the total amount spent by each age group
age_group_spending = df.groupby('Age Group')['Amount'].sum()
plt.figure(figsize=(8, 6))
age_group_spending.plot(kind='bar')
plt.title('Total Amount Spent by Each Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Amount Spent')
plt.show()
```

```python
# 5. Plot a graph depicting the total number of orders from the top 10 states
top_states = df['State'].value_counts().head(10)
plt.figure(figsize=(10, 8))
top_states.plot(kind='bar')
plt.title('Total Number of Orders from Top 10 States')
plt.xlabel('State')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# 6. Determine the total amount spent in the top 10 states
top_states_spending = df[df['State'].isin(top_states.index)].groupby('State')['Amount'].sum()
plt.figure(figsize=(10, 8))
top_states_spending.plot(kind='bar')
plt.title('Total Amount Spent in Top 10 States')
plt.xlabel('State')
plt.ylabel('Total Amount Spent')
plt.show()
```

```python
# 7. Plot a comparison graph between the number of married and unmarried individuals
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Marital_Status')
plt.title('Comparison of Married and Unmarried Individuals')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

```python
# 8. Plot the amount spent by males and females based on marital status
marital_gender_spending = df.groupby(['Marital_Status', 'Gender'])['Amount'].sum().unstack()
plt.figure(figsize=(10, 8))
marital_gender_spending.plot(kind='bar')
plt.title('Amount Spent by Males and Females Based on Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Total Amount Spent')
plt.show()
```

```python
# 9. Plot the count of each occupation present in the dataset
plt.figure(figsize=(12, 10))
sns.countplot(data=df, y='Occupation', order=df['Occupation'].value_counts().index)
plt.title('Count of Each Occupation')
plt.xlabel('Count')
plt.ylabel('Occupation')
plt.show()
```

```python
# 10. Plot the amount spent by each occupation in descending order
occupation_spending = df.groupby('Occupation')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 10))
occupation_spending.plot(kind='bar')
plt.title('Amount Spent by Each Occupation in Descending Order')
plt.xlabel('Occupation')
plt.ylabel('Total Amount Spent')
plt.show()
```

```python
# 11. Provide statistical analysis of each product category based on the percentage of orders completed
product_category_orders = df['Product_Category'].value_counts(normalize=True) * 100
print(product_category_orders)
```

```python
# 12. Determine the budget spent on each product category in descending order
product_category_spending = df.groupby('Product_Category')['Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 10))
product_category_spending.plot(kind='bar')
plt.title('Budget Spent on Each Product Category in Descending Order')
plt.xlabel('Product Category')
plt.ylabel('Total Amount Spent')
plt.show()
```
